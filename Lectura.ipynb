## Estos son los super enhancers

```python

id = data.iloc[[i for i in data.index if  i % 2 ==0],:]
id.reset_index(drop = True, inplace=True)
sec = data.iloc[[i for i in data.index if i % 2 >0],:]
sec.reset_index(drop = True, inplace=True)


enh = []

cant_cromos = 4

for j in range (cant_cromos):

    if j %10000 == 0: print('Tranqui',j)
    if id[0][j][5] == ':':
        cromo = id[0][j][:5]
    else : cromo = id[0][j][:6]

    d = id[0][j][len(cromo)+2:].split('-')
    #print(d)
    ini = int(d[0])
    end= int(d[1])



    s_list = []
    id_list = []

    for i in range(len(sec[0][j][::200])):
        if ini+i*400 < end:
            id_list.append('%s:%i-%i'%(cromo,ini+i*200,ini+i*200+200))
        string = str(sec[0][j][i*200:i*200+200])
        s_list.append(string)
        

    

    enh_ = zip(id_list,s_list)
    
    
    enh.extend(enh_)

    
    #enh_df.columns = ['id','sequence']


print('\n'.join([i for i,j in enh[:10]]))
```

```python
enh_df = pd.DataFrame([[i for i,j in enh],[j for i,j in enh]])

ix  = [i for i,j in enh]
seq = [j for i,j in enh]

tag = [1 for i in range(len(ix))]
enh = np.column_stack((ix,seq, tag))
enh_df = pd.DataFrame(enh)
enh_df.columns=['id', 'sec', 'tag']
enh_df.to_csv('ENH.csv')
```
